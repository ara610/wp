```python
##필요한 라이브러리 임포트

import requests
import pprint
import json
```

```python
#instit_kind = '종합병원' 데이터만 추출
url_1 = "http://apis.data.go.kr/6260000/MedicInstitService/MedicalInstitInfo?serviceKey=gn4AZgDJCNJy8t2gnEAcNpljtl%2BtxqT2KjgpWgNOSW0N7x9IOyAC3bxrZcLTfqbV%2BlvutgkqT1OSjxWR8pUQQA%3D%3D&pageNo=1&numOfRows=10&resultType=json&instit_kind=종합병원"
url_2 = "http://apis.data.go.kr/6260000/MedicInstitService/MedicalInstitInfo?serviceKey=gn4AZgDJCNJy8t2gnEAcNpljtl%2BtxqT2KjgpWgNOSW0N7x9IOyAC3bxrZcLTfqbV%2BlvutgkqT1OSjxWR8pUQQA%3D%3D&pageNo=2&numOfRows=10&resultType=json&instit_kind=종합병원"
```

```python
response = requests.get(url_1)
contents_1 = response.text
```

```python
##보다 예쁘게 출력하기 위해
##indent는 들여쓴다는 의미
pp = pprint.PrettyPrinter(indent=4)
print(pp.pprint(contents_1))
```

```python
json_ob_1 = json.loads(contents_1)
print(json_ob_1)
print(type(json_ob_1))
```

```python
body_1 = json_ob_1['MedicalInstitInfo']['body']['items']['item']
print(body_1)
```

```python
import pandas as pd

#df_1 = 1페이지
df_1 = pd.DataFrame(body_1)
print(df_1)
```

```python
df_1 = df_1[['instit_nm','instit_kind', 'street_nm_addr', 'tel']]
df_1
```

```python
url_2 = "http://apis.data.go.kr/6260000/MedicInstitService/MedicalInstitInfo?serviceKey=gn4AZgDJCNJy8t2gnEAcNpljtl%2BtxqT2KjgpWgNOSW0N7x9IOyAC3bxrZcLTfqbV%2BlvutgkqT1OSjxWR8pUQQA%3D%3D&pageNo=2&numOfRows=10&resultType=json&instit_kind=종합병원"
```

```python
json_ob_2 = json.loads(contents_2)
print(json_ob_2)
print(type(json_ob_2))
```

```python
url_3 = "http://apis.data.go.kr/6260000/MedicInstitService/MedicalInstitInfo?serviceKey=gn4AZgDJCNJy8t2gnEAcNpljtl%2BtxqT2KjgpWgNOSW0N7x9IOyAC3bxrZcLTfqbV%2BlvutgkqT1OSjxWR8pUQQA%3D%3D&pageNo=1&numOfRows=6848&resultType=json"
```

```python
#전체 데이터 추출

response = requests.get(url_3)
contents_3 = response.text
```

```python
pp = pprint.PrettyPrinter(indent=4)
#print(pp.pprint(contents_3))
```

```python
json_ob_3 = json.loads(contents_3)
#print(json_ob_3)
print(type(json_ob_3))
```

```python
body_3 = json_ob_3['MedicalInstitInfo']['body']['items']['item']
#print(body_3)
```

```python
df_3 = pd.DataFrame(body_3)
#print(df_3)
```

```python
df_3 = df_3[['instit_nm','instit_kind', 'street_nm_addr', 'tel']]
#df_3
```

```python
groups = df_3.groupby('instit_kind')
groups.size()
```

```python
df_group = pd.DataFrame(groups.size())
df_group
```

```python
df_group = df_group.reset_index(drop=False, inplace=False)
df_group
```

```python
#컬럼 확인
df_group.columns
```

```python
#컬럼 이름 변경

df_group.columns = ['instit_kind', 'count']
df_group
```

```python
import matplotlib.font_manager as fm

# 설치된 폰트 출력
font_list = [font.name for font in fm.fontManager.ttflist]
font_list
```

```python
# 라이브러리 호출
import matplotlib.pyplot as plt
from matplotlib import rc
#현재 브라우저에 그래프 바로 보여주기
%matplotlib inline
        
rc('font', family='NanumSquareRound') #폰트 지정
plt.rcParams['axes.unicode_minus'] = False #한글 폰트, 마이너스 기호 깨짐 방지
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0'] #원하는 색 지정
plt.title('기관유형',fontsize=10) #그래프 제목 및 폰트 크기 지정
        
w = {"edgecolor": "black", "linewidth":1, "width":0.7} #그래프 테투리 설정: edgecolor(테두리색), linewidth(테투리선의 넓이), width(도넛모양의 파이그래프 생성을 원할 시 그래프의가 표시되는 비율)
plt.pie(df_group['count'], labels = df_group['count'],colors=colors, wedgeprops=w)
plt.legend(df_group['instit_kind']) #데이터 종류 따로 표시
plt.show()
```
