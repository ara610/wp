```python
##필요한 라이브러리 임포트

import requests
import pprint
import json
```

```python
#instit_kind = 종합병원 데이터만 추출
url_1 = "http://apis.data.go.kr/6260000/MedicInstitService/MedicalInstitInfo?serviceKey=gn4AZgDJCNJy8t2gnEAcNpljtl%2BtxqT2KjgpWgNOSW0N7x9IOyAC3bxrZcLTfqbV%2BlvutgkqT1OSjxWR8pUQQA%3D%3D&pageNo=1&numOfRows=10&resultType=json&instit_kind=종합병원"
url_2 = "http://apis.data.go.kr/6260000/MedicInstitService/MedicalInstitInfo?serviceKey=gn4AZgDJCNJy8t2gnEAcNpljtl%2BtxqT2KjgpWgNOSW0N7x9IOyAC3bxrZcLTfqbV%2BlvutgkqT1OSjxWR8pUQQA%3D%3D&pageNo=2&numOfRows=10&resultType=json&instit_kind=종합병원"
```

```python
response = requests.get(url_1)
contents_1 = response.text
```

```python
##보다 예쁘게 출력하기 위해..
##indent는 들여쓴다는 의미
pp = pprint.PrettyPrinter(indent=4)
print(pp.pprint(contents_1))
```

```python
json_ob_1 = json.loads(contents_1)
print(json_ob_1)
print(type(json_ob_1))
```

```python
body_1 = json_ob_1['MedicalInstitInfo']['body']['items']['item']
print(body_1)
```

```python
import pandas as pd

#df_1 = 1페이지
df_1 = pd.DataFrame(body_1)
print(df_1)
```

```python
df_1 = df_1[['instit_nm','instit_kind', 'street_nm_addr', 'tel']]
df_1
```

```python
url_2 = "http://apis.data.go.kr/6260000/MedicInstitService/MedicalInstitInfo?serviceKey=gn4AZgDJCNJy8t2gnEAcNpljtl%2BtxqT2KjgpWgNOSW0N7x9IOyAC3bxrZcLTfqbV%2BlvutgkqT1OSjxWR8pUQQA%3D%3D&pageNo=2&numOfRows=10&resultType=json&instit_kind=종합병원"
```

```python
#2페이지
response = requests.get(url_2)
contents_2 = response.text
```

```python
pp = pprint.PrettyPrinter(indent=4)
print(pp.pprint(contents_2))
```

```python
json_ob_2 = json.loads(contents_2)
print(json_ob_2)
print(type(json_ob_2))
```

```python
body_2 = json_ob_2['MedicalInstitInfo']['body']['items']['item']
print(body_2)
```

```python
df_2 = pd.DataFrame(body_2)
print(df_2)
```

```python
df_2 = df_2[['instit_nm','instit_kind', 'street_nm_addr', 'tel']]
df_2
```

```python
#df 1.2 위,아래 방향으로 연결
total = pd.concat([df_1,df_2])
total
```

```python
#인덱스 초기화 (drop = True인 경우 열이 밖으로 꺼내지는게 아니라 완전히 삭제)

total = total.reset_index(drop=True, inplace=False)
total
```

```python
#df instit_nm 오름차순 정렬 = 병원이름 가나다순 정렬

total.sort_values('instit_nm', ascending=True)
```
