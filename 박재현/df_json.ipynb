```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import requests
import json
import seaborn as sns

url = "https://open.neis.go.kr/hub/schoolInfo"

params = {
    "ATPT_OFCDC_SC_CODE" : "C10",
    "KEY" : "c6f34188b926429998507e1c6bc14aab",
    "Type" : "json",
    "pIndex" : "1",
    "pSize" : "100",
}
headers = {}

response = requests.request("GET", url, headers=headers, params=params)

datasets = json.loads(response.text)
#print(datasets["schoolInfo"][1])
df = pd.DataFrame(datasets["schoolInfo"][1]["row"])

df.head()
```

```python
plt.figure(figsize=(20,5))
plt.rcParams['font.family'] = 'NanumGothic.ttf'

df_s = df[["SCHUL_KND_SC_NM"]].groupby(["SCHUL_KND_SC_NM"])["SCHUL_KND_SC_NM"].count().reset_index(name="count")
sns.barplot(data=df_s, x="SCHUL_KND_SC_NM", y="count")
```
