```python
data.drop(['first','last','street','city','state','zip','trans_num','unix_time','job','merchant'], axis=1, inplace= True)
```

trans_date_trans_time 이 object가 아닌 datatime64[ns]로 변경된 것을 확인할 수 있다.

```python
amt_info = data.groupby('cc_num').agg(['mean', 'std'])['amt'].reset_index()
```

```python
amt_info = data.groupby('cc_num').mean(numeric_only=True)
amt_info = amt_info.drop(['lat','long', 'city_pop', 'merch_lat', 'merch_long', 'is_fraud'], axis=1)
amt_info_std = data.groupby('cc_num').std(numeric_only=True).drop(['lat','long', 'city_pop', 'merch_lat', 'merch_long', 'is_fraud'], axis=1)

amt_info['std'] = amt_info_std['amt']
amt_info['mean'] = amt_info['amt']
amt_info = amt_info.drop('amt', axis=1)

amt_info.head()
```

```python
data = data.merge(amt_info, on='cc_num', how='left')

data
```

```python
# data = data.drop(['std_x', 'std_y', 'mean_x', 'mean_y'], axis=1)

# data
```

이제 기존 데이터에 특정값(amt), 평균(mean), 표준편차(std)가 모두 모였으니 Z 점수를 계산.

z_score = 특정값 - 평균 / 표준편차<br/>

특정값이 정규분포 범위에서 어느 수준에 위치하는지를 나타낸다. (z-점수는 자료가 평균으로부터 표준편차의 몇 배만큼 떨어져 있는지를 보여줍니다.)

```python
data['amt_z_score'] = (data['amt'] - data['mean']) / data['std'] # z-score 계산
```

```python
data[['amt', 'mean', 'std', 'amt_z_score']].head()
```

index 2행을 확인하면 평균 거래 금액이 약 \$69.9 수준인데 해당 지출금액은 \$220로 상당한 차이를 보인다.

```python
data['merch_coord'] = pd.Series(zip(data['merch_lat'], data['merch_long']))

data['cust_coord'] = pd.Series(zip(data['lat'], data['long']))

data
```

```python
data['distance'] = data.apply(lambda x : geopy.distance.distance(x['merch_coord'], x['cust_coord']).km, axis=1)
```

```python
data
```

```python
data
```

```python
distance_info = data.groupby(['cc_num']).mean(numeric_only=True)
distance_info = distance_info.drop(['lat', 'long', 'city_pop', 'merch_lat', 'merch_long', 'is_fraud', 'amt_z_score', 'cat_z_score', 'distance'], axis=1)
distance_info_std = data.groupby(['cc_num']).std(numeric_only=True).drop(['lat','long', 'city_pop', 'merch_lat', 'merch_long', 'is_fraud', 'amt_z_score', 'cat_z_score', 'distance'], axis=1)

distance_info['std'] = distance_info_std['amt']
distance_info['mean'] = distance_info['amt']
distance_info = distance_info.drop('amt', axis=1)

distance_info.head()
```

```python
data = data.merge(distance_info, on='cc_num', how='left')
data['distance_z_score'] = (data['distance']-data['mean'])/data['std']
data.drop(['mean', 'std'], axis=1, inplace=True)
data.head()
```

```python
from sklearn.model_selection import RandomizedSearchCV

params = {
    'n_estimators' : [100, 500, 1000],
    'learning_rate' : [0.01, 0.05, 0.1, 0.3],
    'lambda_l1': [0, 10, 20, 30, 50],
    'lambda_l2': [0, 10, 20, 30, 50],
    'max_depth': [5, 10, 15, 20],
    'subsample': [0.6, 0.8, 1]
}
```
