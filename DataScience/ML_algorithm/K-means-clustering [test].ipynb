```python
file_url = "https://raw.githubusercontent.com/musthave-ML10/data_source/main/customer.csv"
customer = pd.read_csv(file_url)
```

```python
customer.head()
```

```python
customer['cc_num'].nunique()
```

```python
customer['category'].nunique()
```

```python
customer_dummy = pd.get_dummies(customer, columns = ['category'])
```

```python
customer_dummy.head()
```

ë²”ì£¼ë³„ë¡œ ì–¼ë§ˆë§Œí¼ì˜ ê¸ˆì•¡ì„ ì¼ëŠ”ì§€ë¥¼ ê³„ì‚°í•˜ê¸° ìœ„í•´, ë”ë¯¸ ë³€ìˆ˜ë¡œ ë³€í™˜ëœ ì˜ì—­ì— amt(ê±°ëž˜ ê¸ˆì•¡) ë³€ìˆ˜ì˜ ê°’ì„ ê³±í•˜ì—¬ ê³„ì‚°.<br/>


```python
cat_list = customer_dummy.columns[2:] # ë³€ìˆ˜ ì´ë¦„ ë¦¬ìŠ¤íŠ¸ ìƒì„±
for i in cat_list:
    customer_dummy[i] = customer_dummy[i] * customer_dummy['amt']
```

```python
customer_dummy
```

```python
customer_agg = customer_dummy.groupby('cc_num').sum()
```

```python
customer_agg.head()
```

```python
import sklearn

scaler = sklearn.preprocessing.StandardScaler()
scaler_df = pd.DataFrame(scaler.fit_transform(customer_agg), columns=customer_agg.columns, index=customer_agg.index)
```

### 12.5 ê³ ê° ë°ì´í„° ëª¨ë¸ë§ ë° ì‹¤ë£¨ì—£ ê³„ìˆ˜

```python
distance = []
for k in range(2, 10):
    k_model = KMeans(n_clusters=k)
    k_model.fit(scaler_df)
    labels = k_model.predict(scaler_df)
    distance.append(k_model.inertia_)

sns.lineplot(x=range(2, 10), y=distance)
```

ì–´ëŠ í•œ ì§€ì ì—ì„œ í¬ê²Œ ë–¨ì–´ì§€ì§€ ì•Šê³  ë¹„êµì  ì™„ë§Œí•˜ê²Œ ê·¸ëž˜í”„ê°€ ë‚´ë ¤ê°€ê³  ìžˆìŠµë‹ˆë‹¤. Kê°’ì„ ê²°ì •í•˜ê¸°ê°€ ìƒë‹¹ížˆ ì–´ë ¤ìš´ ëª¨ì–‘ìƒˆ.<br/>

ë°˜ë©´, ì‹¤ë£¨ì—£ ê³„ìˆ˜ëŠ” í´ëŸ¬ìŠ¤í„° ë‚´ë¶€ì—ì„œì˜ í‰ê·  ê±°ë¦¬ì™€ ìµœê·¼ì ‘í•œ ë‹¤ë¥¸ í´ëŸ¬ìŠ¤í„° ë°ì´í„°ì™€ í‰ê·  ê±°ë¦¬ë„ ì ìˆ˜ì— ë°˜ì˜!

```python
from sklearn.metrics import silhouette_score

silhouette = []
for k in range(2, 10):
    k_model = KMeans(n_clusters=k)
    k_model.fit(scaler_df)
    labels = k_model.predict(scaler_df)
    silhouette.append(silhouette_score(scaler_df, labels))
```

```python
sns.lineplot(x=range(2, 10), y=silhouette)
```
