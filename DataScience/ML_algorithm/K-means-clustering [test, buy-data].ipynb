### 12.3 데이터 불러오기 및 데이터 확인하기 : 고객 데이터셋

```python
file_url = "https://raw.githubusercontent.com/musthave-ML10/data_source/main/customer.csv"
customer = pd.read_csv(file_url)
```

```python
customer.head()
```

```python
customer['cc_num'].nunique()
```

```python
customer['category'].nunique()
```

```python
customer_dummy = pd.get_dummies(customer, columns = ['category'])
```

```python
customer_dummy.head()
```

범주별로 얼마만큼의 금액을 썼는지를 계산하기 위해, 더미 변수로 변환된 영역에 amt(거래 금액) 변수의 값을 곱하여 계산.<br/>


```python
cat_list = customer_dummy.columns[2:] # 변수 이름 리스트 생성
for i in cat_list:
    customer_dummy[i] = customer_dummy[i] * customer_dummy['amt']
```

```python
customer_dummy
```

```python
customer_agg = customer_dummy.groupby('cc_num').sum()
```

```python
customer_agg.head()
```

```python
import sklearn

scaler = sklearn.preprocessing.StandardScaler()
scaler_df = pd.DataFrame(scaler.fit_transform(customer_agg), columns=customer_agg.columns, index=customer_agg.index)
```

K-Means clustering은 거리 기반 알고리즘이기 때문에 데이터의 스케일에 영향을 받음.

```python
scaler_df
```

### 12.5 고객 데이터 모델링 및 실루엣 계수

```python
distance = []
for k in range(2, 10):
    k_model = KMeans(n_clusters=k)
    k_model.fit(scaler_df)
    labels = k_model.predict(scaler_df)
    distance.append(k_model.inertia_)

sns.lineplot(x=range(2, 10), y=distance)
```

어느 한 지점에서 크게 떨어지지 않고 비교적 완만하게 그래프가 내려가고 있습니다. K값을 결정하기가 상당히 어려운 모양새.<br/>
이는 실제 프로젝트에서 굉장히 빈번하게 나타는 현상으로, 이에 대한 대안으로 실루엣 계수를 사용한다.<br/>
반면, 실루엣 계수는 클러스터 내부에서의 평균 거리와 최근접한 다른 클러스터 데이터와 평균 거리도 점수에 반영!

```python
from sklearn.metrics import silhouette_score

silhouette = []
for k in range(2, 10):
    k_model = KMeans(n_clusters=k)
    k_model.fit(scaler_df)
    labels = k_model.predict(scaler_df)
    silhouette.append(silhouette_score(scaler_df, labels))
```

```python
sns.lineplot(x=range(2, 10), y=silhouette)
```

실루엣 계수는 높은 값일수록 더 좋은 분류를 의미. 따라서 여기에서는 명확하게 k=4일 때 가장 좋은 분류 성능을 낸다.<br/>
실루엣 계수의 단점은 계산 비용이 상대적으로 크다. 따라서, 기본적으로는 엘보우 기법을 우선 활용하고 K값을 찾기 어려울 때 실루엣 계수를 사용한다.

### K-평균 군집화

#### K-means 클러스터링 - 고객 구매 데이터 분석

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import preprocessing
import numpy as np

data = pd.read_excel('./data/CustomerDataSet.xls')

processed_data = data.copy()
data.head()
```

```python
scaler = preprocessing.MinMaxScaler()

processed_data[['ItemsBought', 'ItemsReturned']] = scaler.fit_transform(processed_data[['ItemsBought', 'ItemsReturned']])

plt.figure(figsize=(10,6))
```

```python
for i in range(1, 7):
    # 클러스터 생성
    estimator = KMeans(n_clusters=i)
    ids = estimator.fit_predict(processed_data[['ItemsBought', 'ItemsReturned']])
    plt.subplot(3,2,i)
    plt.tight_layout()
    plt.title("k_value = {}".format(i))
    plt.xlabel("ItemsBought")
    plt.ylabel("ItemsReturned")
    plt.scatter(processed_data['ItemsBought'], processed_data['ItemsReturned'], c=ids)

plt.show()
```

해석해보면, K가 3이나 4일 때 Clustering이 잘 되는 것을 확인할 수 있다. <br/>
구분을 하자면 제품을 많이 사고 유지하는 고객군 - <br/>
제품을 많이 사나 환불을 조금하는 고객군 - <br/>
제품을 많이 안사고 환불도 많이 하는 고객군으로 나눌 수 있다.

### 각 군집과 제품 ID의 관계를 살펴보자

```python
estimator = KMeans(n_clusters=3)

cluster_ids = estimator.fit_predict(processed_data[['ItemsBought', 'ItemsReturned']])

plt.scatter(processed_data['ItemsBought'], processed_data['ItemsReturned'], c=cluster_ids)

for index, c_id, bought, returned, zip_code, product in processed_data.itertuples():
    plt.annotate("clu: {}: {}".format(cluster_ids[index], product), (bought, returned))

plt.xlabel("ItemsBought")
plt.ylabel("ItemsReturned")
plt.show()
```

Clu1인 제품 2435이 많이 안팔릴 뿐만 아니라 환불이 발생하는 것을 알 수 있다. 현재 그래프 상에는 데이터가 정규화되었기 때문에 면밀히 파악하기 어려우니 원래 데이터로 돌아가서 해당 제품을 찾아보자!

### 각 군집과 지역의 관계를 살펴보자

```python
# 플로팅하기
plt.scatter(data['ItemsBought'], data['ItemsReturned'], c=cluster_ids)

# 우편번호로 범례달기
for (index, c_id, bought, returned, zip_code, product) in data.itertuples():
    plt.annotate(zip_code,(bought + 0.6, returned + 0.6))
    
plt.xlabel('ItemsBought')
plt.ylabel('ItemsReturned')

plt.show()
```
