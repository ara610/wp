```python
# 데이터 전처리

df = df.drop(["Ticket", "Cabin"], axis=1)

print(df.shape)
```

```python
# 이름의 칭호만 추출하여 Title에 저장
df["Title"]  =  df.Name.str.extract("([A-Za-z]+)\.", expand = False)

print(df["Title"])
print("\n")
print(df.shape)
```

```python
# 빈도표 만들기
pd.crosstab(df["Title"], df["Gender"]) 

# pandas.crosstab(index, columns, values=None, ...)
# crosstab은 범주형 변수를 기준으로 개수 파악이나
# 수치형 데이터를 넣어 계산할 때 사용한다.
pd.crosstab(index=[df["Gender"], df["Pclass"]], columns=df["Survived"])
```

```python
df["Title"] = df["Title"].replace(["Capt", "Col", "Don", "Jonkheer", "Major", "Rev", "Sir"], "MaleRare")
# 훈련 자료에서 "Capt", "Col", "Don", "Jonkheer", "Major", "Rev", "Sir" 칭호를 "MaleRare"로 교체
df["Title"] = df["Title"].replace(["Countess", "Lady", "Mlle", "Mme", "Ms"], "FemaleRare")
df["Title"] = df["Title"].replace(["Dr"], "Rare")

df

# 드문 칭호를 바꾸기
```

```python
# 훈련 자료에서 칭호에 따른 생존율
df[["Title", "Survived"]] \
.groupby(["Title"], as_index=False) \
.mean() \
.sort_values(by="Survived", ascending=False)
```

```python
# 성별을 숫자로 변환
gender_mapping = {"male": 0, "female": 1}

df["Gender"] = df["Gender"].map(gender_mapping).astype(int)

#칭호를 숫자로 변환

# 각 칭호에 매칭할 숫자
title_mapping = {
    "FemaleRare": 1, 
    "Mrs": 2, 
    "Miss": 3, 
    "Master": 4, 
    "Rare": 5, 
    "MaleRare": 6, 
    "Mr": 7
} 

df["Title"] = df["Title"].map(title_mapping)
df["Title"] = df["Title"].fillna(0) # NaN (Null) 값을 0으로 채움

df["Title"]
```

```python
# 성별, 티켓 등급별 연령 분포 확인
g = sns.FacetGrid(
    df, 
    row="Gender", 
    col="Pclass", 
    hue="Gender", 
    height=2.2,
    aspect=1.6
)
g.map(plt.hist, "Age", bins=20)
g.add_legend() # 범례 추가
```

```python
# 훈련 자료에서 성별, 티켓 등급별 연령 중앙값 계산
import numpy as np



guess_ages = np.zeros((2, 3))
# numpy의 zeros 메소드는 0으로만 채워진 array를 생성합니다.
# 위의 예시는 2차원 array 이며, 행의 개수가 2개 / 열의 개수가 3개 입니다.

# guess_ages.shape : (2, 3) 을 반환하며 guess_ages.shape[0]은 2 입니다.
for i in range(guess_ages.shape[0]): 
    for j in range(guess_ages.shape[1]):
        guess_df = df[(df["Gender"] == i) & (df["Pclass"] == j + 1)]["Age"].dropna()
        # 훈련 자료에서 성별이 i 이고 티켓 등급이 j+1 인 연령 추출하고 결측은 제외
        age_guess = guess_df.median() # 추출한 연령 집단의 중앙값
        # 소수점 반올림해서 연령 추정 표 채우기
        guess_ages[i, j] = round(age_guess)

# 각 티켓 등급별로 중앙값을 비교해보면, 남성이 여성보다 연령이 높음
# 티켓 등급이 좋을수록 연령 중앙값이 높음
        
print(guess_ages)
```

```python
for i in range(guess_ages.shape[0]):
    for j in range(guess_ages.shape[1]):
        # 훈련 자료에서 연령이 결측, 성별이 i, 티켓 등급이 j+1인 샘플의 
        # 연령에 해당하는 추정된 연령 채우기
        df.loc[(df["Age"].isnull()) 
                     & (df["Gender"] == i) 
                     & (df["Pclass"] == j + 1), "Age"] = guess_ages[i, j]
        
        
df["Age"] = df["Age"].astype(int) # 연령을 정수형으로 변환

df.info()
# 연령 결측값이 채워짐을 확인
```
